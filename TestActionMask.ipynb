```python
import random
from typing import Any, Optional

from gymnasium import Env
from gymnasium.spaces import Discrete, Box, Dict, MultiBinary
from numpy import ndarray
from ray.rllib.algorithms.ppo import PPOConfig
from ray.rllib.core.rl_module.rl_module import RLModuleSpec
from TorchDictActMaskPPORLModule import MaskedTorchPPORLModule, MaskedTorchPPORLModuleCatalog
from ray.rllib.examples.envs.classes.action_mask_env import ActionMaskEnv
from testactionmask import TestActionMask
import argparse

%env PYTHONWARNINGS ignore:DeprecationWarning

def train_masked_algorithm(env):
    """
    Train a model for the random environment using the PPO algorithm with default parameters.
    """

    """ Build an RL module spec.  Reference the catalog_class in the spec
        to avoid the normal catalog processing, which does not recognize
        the use of dictionary in the environment observation space
    """
    module_spec = RLModuleSpec(module_class=MaskedTorchPPORLModule, 
                               catalog_class=MaskedTorchPPORLModuleCatalog,
                               observation_space=env.observation_space, 
                               action_space=env.action_space)
   
    base_config = (
        PPOConfig()
        .environment(env=TestActionMask)
        .rl_module(rl_module_spec = module_spec)
        .training(num_epochs = 1)
    )

    # build the configuration, generating an algorithm
    algo = base_config.build()

    # perform training
    result = algo.train()
    print(result)


if __name__ == "__main__":

    # instantiate the environment to gain access to the observation_space and action_space
    env = TestActionMask(None)
    obs,_ = env.reset()

    # train
    train_masked_algorithm(env)
```
